<b>First making a model for binary classification:</b>

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def htheta(theta,x):
    return sigmoid(np.dot(x,theta))
```

```python
def cal_cost(x,y,theta):
    h = htheta(theta,x)
    m = len(x)
    return (-1/m * (np.sum(y*np.log(h)+(1-y)*np.log(1-h))))
```

```python
def gd(alpha,x,y,iterations):
    m = len(x)
    ones =np.ones((m,1)) 
    x = np.concatenate((ones, x), axis=1) 
    n = x.shape[1]
    theta = np.ones(x.shape[1])
    h = htheta(theta,x)
    cost = np.ones(iterations)
    for i in range(0,iterations):
        for j in range(0,n):
            theta[j] = theta[j] - ((alpha/m)*np.sum((h-y)*x[:,j]))   
    return theta
```

```python
def log_reg(x,y,alpha,iterations):
    theta = gd(alpha,x,y,iterations)
    return theta
```

<b>For multiclass classification(One vs all method):</b>

```python
#Create different output classes for each of the output
classes = Y[63].unique()
classes
```

```python
Y_new = np.zeros(shape=(len(X),len(classes)))
Y_new = pd.DataFrame(Y_new,columns=classes)       
Y_new
```

```python
for i in range(0, len(Y)):
    val = Y[63].iloc[i]
    Y_new[val].iloc[i] = 1
        
Y_new
```

```python
theta_all_classes = {}
for i in range(0,len(classes)):
    t = log_reg(np.array(X),np.array(Y_new[classes[i]]).reshape(len(Y),),0.0002,10)
    theta_all_classes[classes[i]] = t
theta_all_classes
```

```python
pd.DataFrame(theta_all_classes).iloc[:,0]
```

```python
predict_all = []
for i in range(0, len(classes)):
    theta_i = pd.DataFrame(theta_all_classes)
    h = htheta(theta_i.iloc[:,i], X)
    predict_all.append(h)
predict_all=pd.DataFrame(predict_all)
```
