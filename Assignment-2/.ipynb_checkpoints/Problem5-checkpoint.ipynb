```python
import pandas as pd
import numpy as np
training = pd.read_csv('optdigits.tra',names=np.arange(64))
df = training.drop(63,axis=1)
```

```python
y = training[[63]]
y
```

```python
def hypothesis(theta, X):
    return 1 / (1 + np.exp(-(np.dot(theta, X.T)))) - 0.0000001
```

```python
def cost(X, y, theta):
    y1 = hypothesis(X, theta)
    return -(1/len(X)) * np.sum(y*np.log(y1) + (1-y)*np.log(1-y1))
```

```python
print(len(df))
print(len(y))
X = pd.concat([pd.Series(1, index=df.index, name='00'), df], axis=1)
```

```python
y = y.iloc[:, 0]
```

```python
classes = y.unique()
```

```python
y1 = np.zeros([df.shape[0], len(classes)])
y1 = pd.DataFrame(y1,columns=classes)
y1
```

```python
y1[0].iloc[5]
```

```python
for i in range(0, len(y)):
    val = y[0].iloc[i]
    y1[val].iloc[i] = 1
        
y1
```

```python
def gradient_descent(X, y, theta, alpha, epochs):
    m = len(X)
    for i in range(0, epochs):
        for j in range(0, 10):
            theta = pd.DataFrame(theta)
            h = hypothesis(theta.iloc[:,j], X)
            for k in range(0, theta.shape[0]):
                theta.iloc[k, j] -= (alpha/m) * np.sum((h-y.iloc[:, j])*X.iloc[:, k])
            theta = pd.DataFrame(theta)
    return theta, cost
```

```python
theta = np.zeros([df.shape[1]+1, y1.shape[1]])
theta = gradient_descent(X, y1, theta, 0.02, 1500)
```
